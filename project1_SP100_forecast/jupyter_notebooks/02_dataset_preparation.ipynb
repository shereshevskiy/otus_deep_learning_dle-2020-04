# <center>Подготовка данных (индекс S&P100)
Подготовим данные для задачи.   
Данные будут представлять из себя многомерный временной ряд из курсов акций, входящих в индекс. Здесь и далее работать будем с индексом **S&P100** (тикер **OEX**). Для прогнозирования везде будем брать цену закрытия (**"Close"**)

```python
import os
os.chdir("..")

%load_ext autoreload
%autoreload 2

from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd
import json
```

```python
path_to_prices = 'prices'
path_to_data = 'data'
```

список акций индекса

```python
with open("sp100_tickers.json", "r") as f:
    sp100_tickers = json.load(f)
print(sp100_tickers)
len(sp100_tickers)
```

```python
data = pd.DataFrame()
for ticker in sp100_tickers:
    df = pd.read_csv(os.path.join(path_to_prices, f"{ticker}.csv"), index_col=0)
    df.rename({"Close": ticker}, axis=1, inplace=True)
    data = pd.concat([data, pd.DataFrame(df[ticker])], axis=1, sort=False)
```

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```

Добавим целевой столбец - индекс **S&P100** (тикер **OEX**).   

```python
ticker = "OEX"
df = pd.read_csv(os.path.join(path_to_prices, f"{ticker}.csv"), index_col=0)
df = df.loc[data.index]  # согласуем временной период (у s&p100 он оказался больше)
df.rename({"Close": "S&P100"}, axis=1, inplace=True)
data = pd.concat([data, pd.DataFrame(df["S&P100"])], axis=1, sort=False)

data.head()
```

```python
data.shape
```

Также видим, что у части акций нет дальних котировок (по-видимому, они начали торговаться позже). Заменим пропуски первыми появившимися котировками (с этим тоже можно поработать, но пока так).

```python
data.fillna(method='backfill', inplace=True)
data.head()
```

Сохраним этот файл и будем дальше работать уже с ним

```python
filename = "sp100.csv"
data.to_csv(os.path.join(path_to_data, filename))
```
