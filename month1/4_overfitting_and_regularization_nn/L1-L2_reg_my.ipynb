```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def l1_reg(x, lambda_=1):
    return lambda_ * abs(x)

def l2_reg(x, lambda_=1):
    return lambda_ * x**2
```

```python
x = np.linspace(-1, 1, 200)
```

```python
functional = lambda x: 0.7*(x - 0.5)**2 - 0.5
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(x, l1_reg(x), label="l1_reg")
ax1.plot(x, functional(x), label="functional")
ax1.plot(x, functional(x) + l1_reg(x), label="functional + l1_reg")
ax1.set_ylim(-1, 1)
ax1.legend()
ax1.plot([0, 0], [-1, 1])
ax1.set_title("L1 regularization")

ax2.plot(x, l2_reg(x), label="l2_reg")
ax2.plot(x, functional(x), label="functional")
ax2.plot(x, functional(x) + l2_reg(x), label="functional + l2_reg")
ax2.set_title("L2 regularization")
ax2.set_ylim(-1, 1)
ax2.legend()
ax2.plot([0, 0], [-1, 1]);

```
